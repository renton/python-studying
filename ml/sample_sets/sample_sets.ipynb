## Wave Dataset

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
X, y
```

```python
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.show()
```
